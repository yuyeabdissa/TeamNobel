```python
data = pd.read_csv('nobelPrize.csv')
```

```python
# Dataset has about 969 rows and 18 coulms, which has details about the nobel prize winners name, birth date, country , and prize details.
data.info()
```

```python
# Repeated Nobel Prize Winners
# Finding the laurates who got nobel price more than once
data.groupby("Full Name").filter(lambda x: len(x) >= 2)['Full Name'].value_counts()
```

```python
## Typical Age of Getting the Prize

data['Birth Date'] = pd.to_datetime(data['Birth Date'],errors='coerce')

data['Age'] = data['Year'] - data['Birth Date'].dt.year
sns.lmplot(x='Year', y='Age', data=data, lowess=True, 
           aspect=2, line_kws={'Color' : 'black'})
```
